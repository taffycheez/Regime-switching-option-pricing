## Extracting data

First we need to get some data. We will use the `yfinance` library and retrieve S&P data from the past 5 years. 

```python
# Importing libraries (you will need to install these if you haven't already 
# using pip install pandas numpy yfinance matplotlib)

# You need to run this again every time you close + reopen the notebook

import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Retrieving data
sp500 = yf.Ticker("^GSPC").history(period="5y")

# Let's take a look 

print(sp500.tail())

# Dropping 'Dividends' and 'Stock Splits'
sp500 = sp500.drop(columns=['Dividends', 'Stock Splits'])

# Summary statistics

sp500.describe()
```
